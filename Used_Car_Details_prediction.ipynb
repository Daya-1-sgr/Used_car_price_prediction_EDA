```python
def extract_manufacturer_model_and_variant(name):
  parts=name.split()
  manufacturer=parts[0]
  model=parts[1]
  variant=' '.join(parts[2:])
  return manufacturer,model,variant
```

```python
df[['manufacturer','model','variant']]=df['name'].apply(lambda x:pd.Series(extract_manufacturer_model_and_variant(x)))
df.head()
```

Dzire is a model,so cahnging it from variant to model

```python
def adjust_model_variant(row):
    if row['variant'].startswith("Dzire"):
        row['model'] = row['model'] + " Dzire"
        row['variant'] = row['variant'].replace("Dzire", "").strip()
    if row['variant'].startswith("R "):
        row['model'] = row['model'] + " R"
        row['variant'] = row['variant'].replace("R", "").strip()
    if row['variant'].startswith("Altis"):
        row['model'] = row['model'] + " Altis"
        row['variant'] = row['variant'].replace("Altis", "").strip()
    if row['variant'].startswith("i10"):
        row['model'] = row['model'] + " i10"
        row['variant'] = row['variant'].replace('i10',"").strip()
    if row['variant'].startswith("800"):
        row['model'] = row['model'] + " 800"
        row['variant'] = row['variant'].replace('800',"").strip()
    return row
```

```python
df=df.apply(adjust_model_variant,axis=1)
df.head()
```

Removing fuel type from variants,which is present in some variants

```python
fuel_type=['Diesel','Petrol','CNG','LPG','Electric']
def remove_fuel_from_variant(row):
  for fuel in fuel_type:
    if fuel in row['variant']:
       row['variant'] = row['variant'].replace(fuel, "").strip()
  return row
```

```python
df=df.apply(remove_fuel_from_variant,axis=1)
df.head()
```
